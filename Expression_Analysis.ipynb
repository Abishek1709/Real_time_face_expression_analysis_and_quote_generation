# Title: Real-time Facial Expression Analysis and Quote Generation Based On Your Mode

```python
import cv2
import numpy as np
from deepface import DeepFace
import matplotlib.pyplot as plt
import keyboard
import pandas as pd
import random
```

```python
df = pd.read_excel('quotes.xlsx')
```

```python
# Create a VideoCapture object to access the camera
cap = cv2.VideoCapture(0)

counter = 0
expressions = []
most_common_expression = " "

quote = " "

while True:
    # Capture frame-by-frame from the camera
    ret, frame = cap.read()

    if not ret:
        print("Error capturing frame from the camera.")
        break

    # Perform facial expression analysis on the captured frame
    result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)

    if result is not None:
        dominant_emotion = result[0]['dominant_emotion']
        expression = dominant_emotion.capitalize()
        expressions.append(expression)

        counter += 1
        if counter == 50:
            # Find the most common expression in the first 10 frames
            most_common_expression = max(set(expressions), key=expressions.count)
            
            # Get a random quote for the most common expression
            quotes_column = df[most_common_expression]
            quotes = quotes_column.dropna().tolist()
            if quotes:
                quote = random.choice(quotes)

            counter = 0
            expressions = []   
        # print the predicted Facial Expression and Random Quote based on your Mode
        text = f"Your mode: {most_common_expression}"
        cv2.putText(frame, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (225, 0, 0), 2)

        quote_text = f"Quote: {quote}"
        cv2.putText(frame, quote_text, (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 255, 0), 1)
        
        Exit = " Press Q to EXIT"
        cv2.putText(frame, Exit, (10, 400), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 225), 2)

    cv2.imshow('Camera', frame)
    
    # to Exit the camera and the running operation
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the VideoCapture object and close the windows
cap.release()
cv2.destroyAllWindows()
```
